```python
import numpy as np
import pandas as pd
```

```python
movies1=pd.read_csv('tmdb_5000_movies.csv')
credits1=pd.read_csv('tmdb_5000_credits.csv')
```

```python
mov=movies1.merge(credits1, on='title')
```

```python
mov.head()
```

```python
# genres
# id
# keyword
# title
# overview
# release date
# cast
# crew

mov=mov[['movie_id','title', 'overview','genres','keywords','cast','crew']]
```

```python
mov.head()
```

```python
mov.isnull().sum()
```

```python
mov.dropna(inplace=True)
```

```python
mov.duplicated().sum()
```

```python
mov.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
mov['genres']=mov['genres'].apply(convert)
```

```python
mov.head()
```

```python
mov['keywords']=mov['keywords'].apply(convert)
```

```python
mov.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
mov['cast']=mov['cast'].apply(convert3)
```

```python
mov.head()
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
mov['crew']=mov['crew'].apply(convert_crew)
```
